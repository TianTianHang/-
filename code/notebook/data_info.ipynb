```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# 读取JSON文件
with open('data_realnewslike/data_realnewslike_final_1.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# 将数据转换成DataFrame
df = pd.DataFrame(data)
df
```

提取Latin词组和其出现次数

```python
latin_phrases = df['phrase'].apply(pd.Series).stack().reset_index(drop=True)
latin_counter = latin_phrases.value_counts()
latin_phrases
```

可视化词频

```python
latin_counter.plot(kind='bar',logy=True, rot=45)
plt.xlabel('Latin Phrases')
plt.ylabel('Frequency')
plt.title('Latin Phrases Frequency in the Dataset')
plt.show()
```

画词频的分布直方图

```python
import numpy as np
plt.figure(figsize=(10, 5))
plt.hist(np.log1p(latin_counter), bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Log Frequency')
plt.ylabel('Frequency Count')
plt.title('Distribution of Latin Phrases Frequency')
plt.show()
```
